# Preprocessing

```python
# Copy of the original Dataframe to be modified
df_copy = df.copy()
```

```python
df_copy.head()
```

```python
# Display all rows and columns of a dataframe
pd.set_option('display.max_rows', None, 'display.max_rows', None)
```

```python
# Drop the 3rd, 4th, and 5th columns for "VEHICLE TYPE CODE" and "CONTRIBUTING FACTOR VEHICLE" as most rows are null
df_copy = df_copy.drop(columns=['CONTRIBUTING FACTOR VEHICLE 3', 'CONTRIBUTING FACTOR VEHICLE 4', 'CONTRIBUTING FACTOR VEHICLE 5', 'VEHICLE TYPE CODE 3', 'VEHICLE TYPE CODE 4', 'VEHICLE TYPE CODE 5'], axis=1)

# Convert the 'CRASH DATE' column from string to datetime
df_copy['CRASH DATE'] = df_copy['CRASH DATE'].astype('datetime64[ns]')

# Create a 'CRASH YEAR' and 'CRASH MONTH' column derived from the 'CRASH DATE' column
df_copy['CRASH YEAR'] = pd.to_datetime(df_copy['CRASH DATE'], format='%Y/%m/%d', errors='coerce').dt.year
df_copy['CRASH MONTH'] = pd.to_datetime(df_copy['CRASH DATE'], format='%Y/%m/%d', errors='coerce').dt.month

# Create a 'CRASH HOUR' column derived from the 'CRASH TIME' column
df_copy['CRASH HOUR'] = pd.to_datetime(df_copy['CRASH TIME'], format='%H:%M').dt.hour

# Filter the dataframe to only include data from the last n amount of years
df_copy = df_copy.loc[df_copy['CRASH DATE'] >= '1/1/2023']

# Reorder the columns
df_copy = df_copy[['CRASH DATE', 'CRASH MONTH', 'CRASH YEAR', 'CRASH TIME', 'CRASH HOUR', 'BOROUGH', 'ZIP CODE', 'LATITUDE', 'LONGITUDE', 'LOCATION', 'ON STREET NAME', 'CROSS STREET NAME', 'OFF STREET NAME',
                   'NUMBER OF PERSONS INJURED', 'NUMBER OF PERSONS KILLED', 'NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF PEDESTRIANS KILLED', 'NUMBER OF CYCLIST INJURED',
                   'NUMBER OF CYCLIST KILLED', 'NUMBER OF MOTORIST INJURED', 'NUMBER OF MOTORIST KILLED',
                   'CONTRIBUTING FACTOR VEHICLE 1', 'CONTRIBUTING FACTOR VEHICLE 2', 'VEHICLE TYPE CODE 1', 'VEHICLE TYPE CODE 2']]
```

```python
df_copy.head()
```

```python
df_copy.info()
```

```python
# Cleaning the 'Vehicle Type Code' columns by combining 'like' vehicle descriptions under one
columns = ['VEHICLE TYPE CODE 1', 'VEHICLE TYPE CODE 2']
target_vehicles = ['usps', 'us post', 'fdny', 'fire', 'dump', 'amb', 'AMUBULANCE', 'sedan', 'van', 'TTailer', 'Trail', 'Tow', 'Box', 'E-bi', 'E-sc', 'scho', 'util', 'fork', 'scoot', 'pic', 'garb', 'trac', 'semi', 'delv', 'deliv', 'motorized home', 'flat', 'moped', 'unk', 'pk', 'bus', 'taxi', 'motorcycle', 'motor scoo', 'motorbike', 'post']
revised_vehciles = ['USPS Vehicle', 'USPS Vehicle', 'Fire Truck', 'Fire Truck', 'Dump Truck', 'Ambulance', 'Ambulance', 'Sedan', 'Van', 'Trailer', 'Trailer', 'Tow Truck', 'Box Truck', 'E-Bike', 'E-Scooter', 'School Bus', 'Utility Truck', 'Forklift', 'Scooter', 'Truck', 'Refuse Truck', 'Tractor Truck', 'Tractor Truck', 'Delivery Truck', 'Delivery Truck', 'RV', 'Flatbed Truck', 'Moped', 'Unknown', 'Truck', 'Bus', 'Taxi', 'Motorcycle', 'Motor Scooter', 'Motorcycle', 'USPS Vehicle']

for column in columns:
  for i in range(len(target_vehicles)):
    df_copy.loc[df_copy[column].str.contains(target_vehicles[i], case=False, na=False), column] = revised_vehciles[i]

target_trucks = ['TRUCK', 'truck', 'Pick-up Truck']
for column in columns:
  for truck in target_trucks:
    df_copy.loc[df_copy[column].str.contains(truck, na=False), column] = 'Truck'

print('Before removing rows: ' + str(df_copy.shape))
# Remove rows where a vehicle type occurs < 10 times to eliminate spelling errors caused by human entries
df_copy = df_copy[df_copy['VEHICLE TYPE CODE 2'].map(df_copy['VEHICLE TYPE CODE 2'].value_counts()) > 9]
df_copy = df_copy[df_copy['VEHICLE TYPE CODE 1'].map(df_copy['VEHICLE TYPE CODE 1'].value_counts()) > 9]

print('After removing rows: ' + str(df_copy.shape))
```

```python
# Transform the Vehicle Type Code, Borough, & Zip code columns so each value has its own column with boolean values
vtc1_df = pd.get_dummies(df_copy['VEHICLE TYPE CODE 1'], prefix='VTC1 - ', dtype='float')
vtc2_df = pd.get_dummies(df_copy['VEHICLE TYPE CODE 2'], prefix='VTC2 - ', dtype='float')
borough_df = pd.get_dummies(df_copy['BOROUGH'], prefix='BOROUGH - ', dtype='float')
zc_df = pd.get_dummies(df_copy['ZIP CODE'], prefix='ZC - ', dtype='float')
concat_df = pd.concat([df_copy, vtc1_df, vtc2_df, borough_df, zc_df], axis=1)

# Reset the index and sort the Dataframe by the 'CRASH DATE' and 'CRASH TIME' column
concat_df = concat_df.reset_index(drop=True).sort_values(by=['CRASH DATE', 'CRASH TIME'])
```
